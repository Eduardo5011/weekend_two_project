# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
- leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
import os

class ParkingGarage():            
    def __init__(self,number_of_spaces):
        self.tickets = 0
        self. ticket_status = {}
        self.parking_spaces = [False for x in range(number_of_spaces)]


    def take_ticket(self):
        if self.__is_full():
            return "Lot is full"
        else:
            self.tickets += 1
            self.ticket_status[self.tickets] = False
            self.__take_parking_space()
            return f"Thank you for parking, your ticket number {self.tickets}. Please pay before leaving garage"
            
        

    def pay_for_parking(self, ticket_number):
        if ticket_number not in self.ticket_status:
            return("Not a valid ticket number")
        elif self.ticket_status[ticket_number] == True:
            return "Ticket is already paid"
        else:
            self.ticket_status[ticket_number] = True
            return "Than you for paying, you have 15:00 minutes to leave the garage"
            
        
        
  
    def leave_garage(self, ticket_number):
        if ticket_number not in self.ticket_status:
            return("Not a valid ticket number")
        elif self.ticket_status[ticket_number] == False:
            return "Please pay before leaving"
        else:
            self.__leave_parking_space()
            return "Thank you for your business. Have a great day!"
        
        

    def __is_full(self):
        for space in self.parking_spaces:
            if space == False:
                return False
        return True   

    def __take_parking_space(self):
        for index, space in enumerate (self.parking_spaces):
            if space == False:
                self.parking_spaces[index] = True
                break

    def __leave_parking_space(self):
        for index, space in enumerate (self.parking_spaces):
            if space == True:
                self.parking_spaces[index] = False
                break
            
class UI():
    def __init__(self, garage):
        self.garage = garage
        
    def start(self):
        while True:
            self.clear_terminal()
            print("Welcome to our parking garage")
            response = input("\n 1 Park \n 2 Pay \n 3 Leave \n 4 Exit \nPlease choose an option: ")
            self.clear_terminal()
            if response == "1":
                print(self.garage.take_ticket())
                input("\mPlease enter to continue ")
            elif response == "2":
                self.clear_terminal()
                ticket = ("Please enter your ticket number: ")
                while not ticket.isnumeric():
                    ticket = input("Please enter a number: ")
                    print(self.garage.pay_for_parking(int(ticket)))
                    input("\nPress enter to continue ")
            elif response =="3":
                self.clear_terminal()
                ticket = ("Please enter your ticket number: ")
                while not ticket.isnumeric():
                    ticket = input("Please enter a number: ")
                    print(self.garage.leave_garage(int(ticket)))
                    input("\nPress enter to continue ")
            elif response == "4":
                self.clear_terminal()
                print("Program Exit")
                break
            elif response == "5":
                print(self.garage.tickets)
                print(self.garage.ticket_status)
                print(self.garage.parking_spaces)
                input("Press any key to continue")
            else:
                print("Please enter a valid option")
                
    def clear_terminal(self):
        os.system('cls' if os.name=='int' else 'clear')
        
    
my_garage = ParkingGarage(5)
ui = UI(my_garage)
ui.start()
```
